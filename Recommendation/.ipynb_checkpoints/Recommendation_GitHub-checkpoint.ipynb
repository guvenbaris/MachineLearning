#  3. Simple Recommender

This section will create a recommendation system, but it will be a simple recommendation system, only based on weighted voting averages. For this, we will calculate the weighted ranking, we will use the formula defined below.

Weighted Rating: ((${v}/{v+m}$) . R )+ ($({m}/{v+m})$ . C)

where,

* v is the number of votes for the movie
* m is the minimum votes required to be listed in the chart
* R is the average rating of the movie
* C is the mean vote across the whole report

```python
pd.options.mode.chained_assignment = None
data["genres"] = data["genres"].fillna("[]").apply(literal_eval).apply(lambda x : [i["name"] for i in x ])
data["year"] = pd.to_datetime(data["release_date"],errors='coerce').apply(lambda x : str(x).split("-")[0])
```

```python
s = data.apply(lambda x : pd.Series(x["genres"],dtype = 'object'),axis = 1 ).stack().reset_index(level = 1,drop = True)
s.name = "genre"
```

```python
genre_df = data.drop("genres",axis = 1).join(s)
```

```python
def based_recommendation(genre,percintile = 0.90):
    
    """
    Categories that can be asked for recommendation question: 
    'Animation', 'Comedy', 'Family', 'Adventure', 'Fantasy', 'Romance',
       'Drama', 'Action', 'Crime', 'Thriller', 'Horror', 'History',
       'Science Fiction', 'Mystery', 'War', 'Foreign', nan, 'Music',
       'Documentary', 'Western', 'TV Movie', 'Carousel Productions',
       'Vision View Entertainment', 'Telescene Film Group Productions',
       'Aniplex', 'GoHands', 'BROSTA TV',
       'Mardock Scramble Production Committee', 'Sentai Filmworks',
       'Odyssey Media', 'Pulser Productions', 'Rogue State', 'The Cartel'
       
    """
    
    df = genre_df[genre_df["genre"] == genre]
    
    vote_count = df[df["vote_count"].notnull()]["vote_count"].astype("int")
    vote_average = df[df["vote_average"].notnull()]["vote_average"].astype("int")
    C = np.mean(vote_average)
    m = vote_count.quantile(percintile)
    
    qualified = data[(data['vote_count'] >= m) & (data['vote_count'].notnull()) & (data['vote_average'].notnull())] \
    [['title', 'year', 'vote_count', 'vote_average', 'popularity']]
    
    qualified["vote_count"] = qualified["vote_count"].astype("int")
    qualified["vote_average"] = qualified["vote_average"].astype("int")
    
    qualified['wr'] = qualified.apply(lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average'])+(m/(m+x['vote_count']) * C), axis=1)
    
    qualified = qualified.sort_values('wr', ascending=False).head(250)
    
    return qualified
```

# 4. Content Based Recommendation 

This section we will apply Content Based Recommendation. Content-based filtering uses item features to recommend other items similar to what the user likes, based on their previous actions or explicit feedback. We will use links_small.csv dataset. For content-based recommendation we will use overview and tagline content. We need to clear the text in these columns thus we can get a vector that is not longer for this we will define function. This function will return clean text. We need to vectorizer our text so the computer can understand what the words mean. After we will looks vector for cosine similarity. Last step, we will  define the function and this function will return recommendations. 

```python
link_small = pd.read_csv("C:/Users/Barış/Desktop/BTKAKADEMİ EĞİTİM/Kaggle & Github/DATASET/Movie_Recoomendation/links_small.csv")
link_small = link_small[link_small['tmdbId'].notnull()]['tmdbId'].astype('int')
```

* Next step, we will define stop word and lemmatization method

```python
stop_word = set(stopwords.words('english'))
lemma = WordNetLemmatizer()
nltk.download('wordnet')
```

* Next step, we will define text_cleaning function. This function will return cleaned stop words, cleared of non-alphabetic characters and used lemmatization.

```python
def text_cleaning(text): 
    
    text = re.sub(r"[^a-zA-Z]",' ',str(text))
    
    text = word_tokenize(text)
    
    text = [word.lower() for word in text if word.lower() not in stop_word] 
    
    text = ' '.join([lemma.lemmatize(word = w ,pos = 'v') for w in text])
    
    return text
```

```python
# Three id has different kind values do we took a error.
# if you don't want to take a error drop this id number.
data = data.drop([19730, 29503, 35587]) 
```

```python
data["id"] = data["id"].astype("int")
```

```python
# We get the data corresponding to the id values we have from the meta dateset.
link_small = data[data["id"].isin(link_small)] 
```

```python
link_small["overview"] = link_small["overview"].astype("string")
link_small["tagline"] = link_small["tagline"].astype("string") 

# Combine tagline and overview columns as description column
link_small["description"] = link_small["tagline"]+' '+link_small["overview"]
link_small["description"].fillna(value = '',inplace = True)

link_small["description"] = link_small["description"].apply(text_cleaning)
```

```python
# Show clean text
link_small["description"].head()
```

```python
# Vectorize clean text column(description)
tf = TfidfVectorizer(stop_words= 'english',ngram_range=(1,2),analyzer = 'word',min_df=0,lowercase=False)

tfidf = tf.fit_transform(link_small["description"])
tfidf.shape
```

```python
# find cosine similarity 
cos_sim = linear_kernel(tfidf,tfidf)

link_small = link_small.reset_index()

titles = link_small["title"]

# index with newly created index
indices = pd.Series(link_small.index, index=link_small['title'])
```

```python
# get recommendation result
def content_recommendation(title):
    """
    Give movie title, it will return recommendation result
    """
    idx = indices[title]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_score = sorted(sim_scores, key = lambda x : x[1],reverse = True)
    sim_score = sim_score[1:21]

    movie_indeces = [i[0] for i in sim_score]

    return pd.DataFrame(titles.loc[movie_indeces])
```

#  5. Simple Recommendation And Content  Based Recommendation Combination   

This section we will combine simple recommender and content based reccommendation firstly we will find content based recommendation to the result of this we. After we will use simple recommender thus we can regular our recommendation by weighted rating. In conclusion we will give to for recommendation movies with the same content and the highest weighted average.

```python
# Take year for link_small dataset.
link_small["year"] = pd.to_datetime(link_small["release_date"],errors='coerce').apply(lambda x : str(x).split("-")[0])
```

```python
# For films with the same content and the most weighted average (weighting ratio)
def simple_content_recommendation(title,percintile = 0.55):
    
    idx = indices[title]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_score = sorted(sim_scores, key = lambda x : x[1],reverse = True)
    sim_score = sim_score[1:21]

    movie_indeces = [i[0] for i in sim_score]

    df = link_small.loc[movie_indeces]
    
    vote_count = df[df["vote_count"].notnull()]["vote_count"].astype("int")
    vote_average = df[df["vote_average"].notnull()]["vote_average"].astype("int")
    C = np.mean(vote_average)
    m = vote_count.quantile(percintile)
    
    qualified = df[(df['vote_count'] >= m) & (df['vote_count'].notnull()) & (df['vote_average'].notnull())] \
    [['title', 'vote_count', 'vote_average', 'popularity',"year"]]
    
    qualified["vote_count"] = qualified["vote_count"].astype("int")
    qualified["vote_average"] = qualified["vote_average"].astype("int")
    
    qualified['wr'] = qualified.apply(lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average'])+(m/(m+x['vote_count']) * C), axis=1)
    
    qualified = qualified.sort_values('wr', ascending=False)
    
    return qualified
```

```python
simple_content_recommendation("Toy Story")
```

# 6. Recommendation Engine With Graph

In this section, we'll show you the movie relationships on the network. We will use link_small.csv, keywords.csv and credits.csv for this. We'll show the cast, director, genre, likeness, and movie name on the network. We will find similarities with keywords. 

## 6.1 Data preprocessing

```python
# Read datasets
credit_df = pd.read_csv("C:/Users/Barış/Desktop/BTKAKADEMİ EĞİTİM/Kaggle & Github/DATASET/Movie_Recoomendation/credits.csv")
keyword_df = pd.read_csv("C:/Users/Barış/Desktop/BTKAKADEMİ EĞİTİM/Kaggle & Github/DATASET/Movie_Recoomendation/keywords.csv")
```

```python
# change id type to int
credit_df['id'] = credit_df['id'].astype('int')
keyword_df['id'] = keyword_df['id'].astype('int')
link_small["id"] = link_small['id'].astype('int')
```

```python
# combining data
link_small = link_small.merge(credit_df,on='id')
link_small = link_small.merge(keyword_df,on = 'id')
```

```python
# Take casts and director names 
link_small["casts"] = link_small["cast"].apply(literal_eval).apply(lambda x : [i['name'] for i in x])
link_small["director"] = link_small["crew"].apply(literal_eval).apply(lambda x : [i['name'] for i in x  if i['job'] == 'Director'])
```

```python
# Just take 10 casts otherwise on graph they take lots of place
# if you show all off them don't run this code block
link_small["casts"] = link_small["casts"].apply(lambda x : x[:10])
```

```python
#Take key words
link_small["key_words"] = link_small["keywords"].apply(literal_eval).apply(lambda x : [i["name"] for i in x])
```

```python
# Create new dataset and get necessary columns
data_network = link_small[['id',"title","director","key_words","casts",'genres','popularity']]
```

```python
# Define function for clean the key words from useless words or other thinks
def filter_keywords(text): 
    
    text = re.sub(r"[^a-zA-Z]",' ',str(text))
    
    text = word_tokenize(text)
    
    text = ' '.join([word.lower() for word in text if word.lower() not in stop_word])
    
    return text
```

```python
data_network["key_words"] = data_network["key_words"].apply(filter_keywords)
data_network["popularity"] = data_network["popularity"].astype("float")
```

```python
text_content = data_network["key_words"]

# vectorize key_words
vector = TfidfVectorizer(max_df=0.4,        
                             min_df=1,     
                             stop_words='english', 
                             lowercase=False, 
                             use_idf=True,   
                             norm=u'l2',     
                             smooth_idf=True 
                            )
tf_idf = vector.fit_transform(text_content)
tf_idf.shape
```

## 6.2 Show With Network

```python
# Function for find similarities
def find_similar(tfidf_matrix, index, top_n = 5):
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]
    return [index for index in related_docs_indices][0:top_n]  
```

```python
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [14,14]
```

```python
# Define Network
G = nx.Graph(label="MOVIE")
start_time = time.time()
for i, rowi in data_network.iterrows():
    if (i%1000==0):
        print(" iter {} -- {} seconds --".format(i,time.time() - start_time))
    G.add_node(rowi['title'],key=rowi['id'],label="MOVIE" ,mtype=rowi['genres'],rating=rowi['popularity'])

    for element in rowi['casts']:
        G.add_node(element,label="PERSON")
        G.add_edge(rowi['title'], element, label="ACTED_IN")
        
    for element in rowi['genres']:
        G.add_node(element,label="CAT")
        G.add_edge(rowi['title'], element, label="CAT_IN")
        
    for element in rowi['director']:
        G.add_node(element,label="DIRECTOR")
        G.add_edge(rowi['title'], element, label="DIRECTED")
        
           
    indices = find_similar(tf_idf, i, top_n = 5)
    
    snode="Sim("+rowi['title'][:15].strip()+")"        
    G.add_node(snode,label="SIMILAR")
    G.add_edge(rowi['title'], snode, label="SIMILARITY")
    
    for element in indices:
        G.add_edge(snode, data_network['title'].loc[element], label="SIMILARITY")
print(" finish -- {} seconds --".format(time.time() - start_time))       
```

```python
# Show on network
def get_all_adj_nodes(list_in):
    sub_graph=set()
    for m in list_in:
        sub_graph.add(m)
        for e in G.neighbors(m):        
                sub_graph.add(e)
    return list(sub_graph)
def draw_sub_graph(sub_graph):
    subgraph = G.subgraph(sub_graph)
    colors=[]
    for e in subgraph.nodes():
        if G.nodes[e]['label']=="MOVIE":
            colors.append('blue')
        elif G.nodes[e]['label']=="PERSON":
            colors.append('red')
        elif G.nodes[e]['label']=="CAT":
            colors.append('green')
            
        elif G.nodes[e]['label']=="DIRECTOR":
            colors.append('yellow')
            
        elif G.nodes[e]['label']=="SIMILAR":
            colors.append('orange')    


    nx.draw(subgraph, with_labels=True, font_weight='bold',node_color=colors)
    plt.show()
```

```python
list_in = ["Toy Story","Ted","Monster House"]
sub_graph = get_all_adj_nodes(list_in)
draw_sub_graph(sub_graph)
```

```python
list_in = ["The Godfather","The Godfather: Part II","The Godfather: Part III"]
sub_graph = get_all_adj_nodes(list_in)
draw_sub_graph(sub_graph)
```
