 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">2. Imports  </p>

```python
import os 
import re
import cv2
import numpy as np 
import pandas as pd 
from tqdm import tqdm
import seaborn as sbn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score,f1_score

from keras.utils import to_categorical 
from keras.models import Sequential 
from keras.layers import Dense,Dropout,MaxPooling2D
from keras.layers import Conv2D,BatchNormalization,Flatten
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping
```

```python
# In here you must to define your file path 
path = "D:\\Natural_Images\\"
```

```python
# Let's define file path for each object 
loc0 = path+"airplane"
loc1 = path+"car"
loc2 = path+"cat"
loc3 = path+"dog"
loc4 = path+"flower"
loc5 = path+"fruit"
loc6 = path+"motorbike"
loc7 = path+"person"
```

<a id='3'></a>
 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">3. Visualization  </p>

In this section we will show twelve images from the dataset, these images are completely randomly selected. They can change for each working. Last step(One Picture Prediction) we will use one of them pictures for prediction.

* In the next step, we will import photos from file to list, After we will choose randomly twelve picture.

```python
loc_list = [loc0,loc1,loc2,loc3,loc4,loc5,loc6,loc7] # File pathes
pic_name = [] # We add all pictures in a list.
for loc in loc_list:
    for pic in tqdm(os.listdir(loc)):
        pic_name.append(pic)
        
numbers = np.random.randint(0,6899,12) #it give us random twelve number
pictures = []
for i in numbers: # Choose picture with random numbers from pic_name list
    pictures.append(pic_name[i])
```

* Next step, we will take  each pictures name for write under the picture on graph 

```python
path = "D:\\Natural_Images\\"
category_path = []
for picture in pictures: 
    pic = re.sub("[^a-zA-Z]",' ',picture)
    pic_true = pic.split()
    category = path+pic_true[0]
    category_path.append(category)
```

* In the next step, preprocess  pictures for visualization. Firstly read pictures from list. After
resize the pictures for visualization

```python
visualization_pic = []
for path,pic in zip(category_path,pictures):
    img = cv2.imread(os.path.join(path,pic),3)
    vis_img = cv2.resize(img,(150,150))
    visualization_pic.append(vis_img)
```

* Next step is show us pictures on graph.

```python
num_rows = 4
num_cols = 3
fig,axes = plt.subplots(nrows=num_rows, ncols = num_cols,figsize=(12,16))
st = fig.suptitle("Randomly Pictures From Dataset", fontsize=16)
for index,picture in enumerate(visualization_pic):
    i,j = (index // num_cols, index % num_cols)
    name = category_path[index].split('\\')[2]
    axes[i,j].imshow(picture)
    axes[i,j].set_xlabel(f"Category Name: {name.capitalize()}")
    
st.set_y(0.99) 
fig.subplots_adjust(top=0.55,wspace=10,hspace=10)
fig.tight_layout()
plt.show()
```

<a id='4'></a>
 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">4. Preprocessing  </p>

This section we will prepare photos for training. Firstly we need to read from file for this we need 'for' loop so we can take each picture to list.Also we need to give label for each picture thus we can separate pictures. Then we give the pixel values for each picture. After adding the pictures to list. Transform list to array so we can to obtain array thus we have a shape. This shape (number of variable, width,height,channel) now we just need to scale for features and one more step for labels. We need determine labels so our model can't confuse labels.

* Next step we will prepare features and labels values. Features are ours pictures. Labels are what the pictures are.

```python
loc_list = [loc0,loc1,loc2,loc3,loc4,loc5,loc6,loc7]
features = []
labels = []
for label,loc in enumerate(loc_list):
    for i in tqdm(os.listdir(loc)):
        img_read = cv2.imread(os.path.join(loc,i),3)
        img = cv2.resize(img_read,(50,50))
        features.append(np.array(img))
        labels.append(label)
```

```python
# Total Picture 
print("Total Picture Number:",len(features))
```

```python
# List to Matrix 
X = np.array(features)
X = X/255
y = np.array(labels)
y = to_categorical(y,8)
```

```python
print("Features Shape: ",X.shape)
print("Label Shape:",y.shape)
```

<a id='5'></a>
 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">5. Split Data To Train And Test</p>

In this section we define function and this function split dataset to train and test.

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size =0.2,random_state=54)
```

```python
print("X Train Shape: ",x_train.shape)
print("Y Train Shape: ",y_train.shape)

print("X Test Shape: ",x_test.shape)
print("Y Test Shape: ",y_test.shape)
```

<a id='6'></a>
 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">6. Model </p>

This section we will create convolutional neural network(CNN) model. Firstly we definition params to will use.  Secondly we will create CNN model with keras library.

```python
# Model params 
shape = (50,50,3)
epoch = 20
batch_size = 16
```

```python
model = Sequential([
    Conv2D(32,(3,3),input_shape =shape,activation = 'relu',kernel_initializer='normal'),
    MaxPooling2D(pool_size=(2,2)), 
    BatchNormalization(),
    
    Conv2D(64,(3,3),activation='relu',padding = 'same'),
    BatchNormalization(), 

    Conv2D(128,(3,3),padding ='same',activation="relu"),
    MaxPooling2D(pool_size=(2,2)),
    BatchNormalization(), 
    
    Conv2D(256,(3,3),padding='same',activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    
    Flatten(),
    Dense(256,activation='relu'),
    Dropout(0.4),
    Dense(8,activation = 'softmax')
    
])
model.summary()
```

```python
model.compile(loss = "categorical_crossentropy",optimizer ="Adam",metrics =["accuracy"])

early_stop  = EarlyStopping(mode = 'min',patience = 10,monitor = "val_accuracy",verbose = 1)
```

```python
hist = model.fit(x_train,y_train,epochs = epoch,verbose = 1,batch_size = batch_size, 
                validation_split = 0.1,callbacks = [early_stop])
```

<a id='7'></a>
 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">7. Model Evaluate</p>

This section we evaluate our model. Firstly we will look model loss and model accuracy on graph. After we will look accuracy score, confusion matrix, f1 score. The loss graph tells us whether the model loss is decreasing or not. We can understand model situation because loss must regularly decreasing. The accuracy graph must regularly increasing so we can tell the model was success. After we will test our model with test dataset thus look accuracy score, confusion matrix and f1 score. Thus we make a decision the model success is enough or not. 

```python
%matplotlib inline
plt.plot(hist.history["loss"])
plt.plot(hist.history["val_loss"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Graph")
plt.legend(["train","test"])
plt.show()
```

```python
%matplotlib inline
plt.plot(hist.history["accuracy"])
plt.plot(hist.history["val_accuracy"])
plt.xlabel("Epochs")
plt.ylabel("accuracy")
plt.title("Accuracy Graph")
plt.legend(["train","test"])
plt.show()
```

```python
y_hat = model.predict(x_test)
y_pred = np.argmax(y_hat,axis = 1)
y_true = np.argmax(y_test,axis = 1)
```

```python
cm = confusion_matrix(y_true,y_pred)
plt.figure(figsize=(10,8))
sbn.heatmap(cm,annot = True,fmt = '.1f',cmap = 'RdBu_r',linewidth = "1")
```

```python
acc = accuracy_score(y_true,y_pred)
print("Accuracy Score: %",acc*100)
```

```python
f1 = f1_score(y_true,y_pred,average="macro")
print("F1 Score:",f1)
```

 # <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;">8. One Picture Prediction</p>

In this step we will make our model predict the picture. It will be one of the pictures we show in the picture visualization section that we will use, but we will also randomly select it. After we give the model to predict it. Lastly  we will show result to prediction on screen.

```python
# In this step we will upload image from dataset folders 
# after we will make our model predict the picture but this will happen randomly 
# We will use the random variables we used at the for visualization.
```

```python
value = int(np.random.randint(0,12,1))

img_read_test = cv2.imread(os.path.join(category_path[value],pictures[value]),3)
test_img = cv2.resize(img_read_test,(50,50))
test_image = test_img.reshape(-1,test_img.shape[0],test_img.shape[1],test_img.shape[2])

test_array = (np.array(test_image)/255)

prediction = model.predict(test_array)
prediction_df = pd.DataFrame({'pred':prediction[0]})
```

```python
label_dict = {0:"airplane",1:"car",2:"cat",3:"dog",4:"flower",5:"fruit",6:"motorbike",7:"person"}

for key in label_dict:
    if key == prediction_df.sort_values(by="pred",ascending = False).index[0]:
        print(f"Model prediction is {label_dict.get(key).capitalize()} rate % {format(float(prediction_df.loc[key].values*100),'.5f')}")
        real_name = category_path[value].split('\\')[2]
        print(f"Real picture name: {real_name.capitalize()}, Index no: {value}]")
```
